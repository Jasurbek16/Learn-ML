```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def perceptron(X, y, theta, learning_rate, iterations):
    X_b = np.c_[np.ones((len(X), 1)), X]

    for _ in range(iterations):
        # loop through each training example
        for j in range(len(X_b)):
            x_i = X_b[j:j+1] # get the j-th training example
            y_i = y[j] # get the corresponding label
            y_pred = 1 if np.matmul(x_i, theta) >= 0 else 0  # compute the predicted output using the step function
            # update the weights if the prediction is incorrect          
            theta = theta + learning_rate * (y_i - y_pred) * x_i.T if y_pred != y_i else theta

    return theta    
```

```python
np.random.seed(42)

X = np.random.randn(100, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int).reshape(-1, 1)

plt.scatter(X[:, 0], y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Training Data')
plt.show()
```

```python
theta = np.zeros((3, 1))
theta = perceptron(X, y, theta, learning_rate=0.01, iterations=1000)

print(theta)
```

```python
X_b = np.c_[np.ones((len(X), 1)), X]
y_pred = (np.matmul(X_b, theta) >= 0).astype(int)
accuracy = np.mean(y_pred == y)
print(f"accuracy: {accuracy * 100:.2f}%")
```
